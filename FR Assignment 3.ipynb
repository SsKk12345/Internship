# Taking user input to search for a product

```python
# importing required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
# connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\sonus\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# sending url to open
driver.get('http://www.amazon.in/')
```

```python
# taking user input to search for product
search_box = driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
a = input('Enter the product to search:')
search_box.send_keys(a)

# clicking to search for the entered product
click = driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
click.click()
```

# Getting the details of the search results

```python
# making lists to add collected data into it
url = []
bn = []
np = []
price = []
re = []
ed = []
av = []
```

```python
# scraping the urls of the product from first 3 pages
for page in range(3):
    
    ur = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in ur:
        url.append(i.get_attribute('href'))
        
    # finding and clicking on next button to go to next page
    nex = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nex.click()
    time.sleep(10)
```

```python
# finding prduct details of all the collected urls
for i in url:
    driver.get(i)   # opening the urls collected one by one 
    time.sleep(2)
    
    try:
        # finding brand name of the product
        brand = driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        bn.append(brand.text)
        
        # finding product name
        name = driver.find_element(By.XPATH,'//*[@id="title_feature_div"]/div/h1/span')
        np.append(name.text)
        
        # finding price
        pr = driver.find_element(By.XPATH,'//*[@id="corePrice_feature_div"]/div/span/span[2]/span[2]')
        price.append(pr.text)
        
        # finding return/exchange data
        rx = driver.find_element(By.XPATH,'//*[@id="icon-farm-widget-0"]/div/div/div/div[2]/span/div[2]/a')
        re.append(rx.text)
        
        # finding expected delivery data
        de = driver.find_element(By.XPATH,'//*[@id="deliveryBlock_feature_div"]/div/div/div[2]/span/span[1]')
        ed.append(de.text)
        
        # finding availability data
        avail = driver.find_element(By.XPATH,'//*[@id="availabilityInsideBuyBox_feature_div"]/div/div[1]/span')
        av.append(avail.text)
        
    # if some of the data is not found putting hyphen in place of that
    except NoSuchElementException:
        bn.append('-')
        np.append('-')
        price.append('-')
        re.append('-')
        ed.append('-')
        av.append('-')
```

```python
# checking the length of data collected
len(url),len(bn),len(np),len(price),len(re),len(ed),len(av)
```

```python
# taking desired length of data to make all data length equal
brand = bn[:174]
product_name = np[:174]
pr = price[:174]
```

```python
# checking again length of collected data after making changes
len(url),len(brand),len(product_name),len(pr),len(re),len(ed),len(av)
```

```python
# making dataframe
df = pd.DataFrame({'Brand':brand,'Name of Product':product_name,'Price':pr,'Return/Exchange':re,'Expected Delivery':ed,'Availability':av,'URL':url})
df
```

```python
# Saving this in csv file
df.to_csv('guitar.csv',index=False,header=True)
```

```python
# making empty lists to add different data in it
url = []
bn = []
pn = []
color = []
ram = []
rom = []
pc = []
sc = []
ds = []
bc = []
pr = []
```

```python
# finding urls of the product
urls = driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in urls:
    url.append(i.get_attribute('href'))
```

```python
# finding brand names and adding them in list
b = driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in b:
    n = i.text.split()
    bn.append(n[0])
```

```python
# finding price
p = driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
for i in p:
    pr.append(i.text)
```

```python
# using for loop to find rest data from url
for i in url:
    driver.get(i)
    time.sleep(2)
    
    # clicking on read more button to scrap all the data
    rm = driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]')
    rm.click()
             
    try:
        # finding product name
        p = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div[1]/div[1]/table/tbody/tr[3]/td[2]/ul/li')
        pn.append(p.text)
        
        # finding color
        c = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div[1]/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        color.append(c.text)
        
        # finding RAM
        ra = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div[1]/div[4]/table/tbody/tr[2]/td[2]/ul/li')
        ram.append(ra.text)
        
        # finding ROM
        ro = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div[1]/div[4]/table/tbody/tr[1]/td[2]/ul/li')
        rom.append(ro.text)
        
        # finding primary camera
        pca = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div[1]/div[5]/table/tbody/tr[1]/td[2]/ul/li')
        pc.append(pca.text)
        
        # finding secondary camera
        sca = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div[1]/div[5]/table/tbody/tr[2]/td[2]/ul/li')
        sc.append(sca.text)
        
        # finding display size
        d = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div[1]/div[2]/table/tbody/tr[1]/td[2]/ul/li')
        ds.append(d.text)
        
        # finding battery capacity
        b = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[7]/div[4]/div/div[2]/div[1]/div[7]/table/tbody/tr/td[2]/ul/li')
        bc.append(b.text)        
        
    # Handling exception
    except NoSuchElementException: 
        # if some of the data is not found then replacing it by hyphen '-'
        pn.append('-')
        color.append('-')
        ram.append('-')
        rom.append('-')
        pc.append('-')
        sc.append('-')
        ds.append('-')
        bc.append('-')
```

# Hostel Details

```python
# importing required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
# connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\sonus\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# sending url to search on the driver
driver.get('https://www.hostelworld.com/')
```

```python
# entering the location
loc = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/div/input')
loc.send_keys('London')
```

```python
# Selecting London location
lo = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div')
lo.click()
```

```python
# clicking to search
go = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[5]')
go.click()
```

```python
# making empty lists to add different data in it
name = []
distance = []
rating = []
t_review = []
o_review = []
pf_price = []
dm_price = []
facility = []
p_desc = []
```

```python
# using for loop to go on next page to scrap all the data
for page in range(5):
    # finding name of hostel
    n = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
    for i in n:
        name.append(i.text)
        
    # finding distance from city center
    d = driver.find_elements(By.XPATH,'//span[@class="description"]')
    for i in d:
        o = i.text.split()
        distance.append(o[2])
        
    # finding rating
    r = driver.find_elements(By.XPATH,'//div[@class="bottom-rating"]/a/div/div[1]')
    for i in r:
        rating.append(i.text)
            
    # finding total reviews
    tre = driver.find_elements(By.XPATH,'//div[@class="bottom-rating"]/a/div/div[2]/div[2]')
    for i in tre:
        t = i.text.split()
        t_review.append(t[0])
            
    # finding overall reviews
    ore = driver.find_elements(By.XPATH,'//div[@class="bottom-rating"]/a/div/div[2]/div[1]/span')
    for i in ore:
        o_review.append(i.text)
            
    # finding private from price
    pfp = driver.find_elements(By.XPATH,'//div[@class="prices-col"]/a[1]/div[1]/div')
    for i in pfp:
        pf_price.append(i.text)
    
    # finding dorms from price
    dpre = driver.find_elements(By.XPATH,'//div[@class="prices-col"]/a[1]/div[2]/div')
    for i in dpre:
        dm_price.append(i.text)
            
    # finding facilities
    fc = driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
    for i in fc:
        facility.append(i.text.split('\n'))
            
    # finding property description
    pds = driver.find_elements(By.XPATH,'//div[@class="rating-factors prop-card-tablet rating-factors small"]')
    for i in pds:
        p_desc.append(i.text.split('\n'))
            
    # clicking on next button to go on next page
    try:
        next_but = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[4]/div/div/div[33]/div[5]')
        next_but.click()
        
    except NoSuchElementException:
        pass
```
