# Most viewed videos on YouTube

```python
# importing required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\sonus\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# sending url to the webdriver
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
# making empty lists to add different data in it
rank = []
name = []
artist = []
u_date = []
views = []
```

```python
# finding the rank of the most viewed videos
ra = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in ra[0:30]:  # scraping only desired data i.e top 30
    rank.append(i.text)
    
# finding video name
n = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for i in n[0:30]:
    name.append(i.text)
    
# finding artist name
art = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for i in art[0:30]:
    artist.append(i.text)
    
# finding video upload date
date = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for i in date[0:30]:
    u_date.append(i.text)
    
# finding views
vw = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for i in vw[0:30]:
    views.append(i.text)
```

```python
# checking the length of data collected
len(rank),len(name),len(artist),len(u_date),len(views)
```

```python
# making dataframe
df = pd.DataFrame({'Rank':rank,'Name':name,'Artist Name':artist,'Upload Date':u_date,'Views in Billions':views})
df
```

# Team India's international fixtures

```python
# importing required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\sonus\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# sending url to the webdriver
driver.get('https://www.bcci.tv/')
```

```python
# clicking on international fixtures
int_fix = driver.find_element(By.XPATH,'/html/body/nav/div[1]/div[2]/ul[1]/li[2]/a')
int_fix.click()
```

```python
# making different lists to append data in it
title = []
series = []
place = []
date = []
time = []
```

```python
# we have to click view more fixtures to find all fixtures
for more in range(10):
    # finding match title
    mt = driver.find_elements(By.XPATH,'//div[@class="fixture-card-bottom"]/div/span[1]')
    for i in mt:
        title.append(i.text.split('-')[0])
        
    # finding series
    sr = driver.find_elements(By.XPATH,'//span[@class="ng-binding"]')
    for i in sr:
        series.append(i.text)
        
    # finding place
    pl = driver.find_elements(By.XPATH,'//div[@class="fix-place ng-binding ng-scope"]')
    for i in pl:
        h = i.text.split('-')
        place.append(h[1])
    
    # finding date
    dt = driver.find_elements(By.XPATH,'//div[@class="match-top-info d-flex align-items-center justify-content-between"]/div[1]/h5')
    for i in dt:
        date.append(i.text)
        
    # finding time
    ti = driver.find_elements(By.XPATH,'//div[@class="match-top-info d-flex align-items-center justify-content-between"]/div[2]')
    for i in ti:
        time.append(i.text)
        
    # clicking on more fixture button
    try:
        mf = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div/div/div/div[2]/div[3]/div[2]/div/button')
        mf.click()
        
    except NoSuchElementException:
        pass
```

```python
# checking length of collected data
len(title),len(series),len(place),len(date),len(time)
```

```python
# making dataframe
df = pd.DataFrame({'Match title':title,'Series':series,'Place':place,'Date':date,'Time':time})
df
```

# Trending repositories on Github

```python
# importing required libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\sonus\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# sending url to the driver
driver.get('https://github.com/')
```

```python
# clicking on toggle option
toggle = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[1]/div[2]/button/span/span')
toggle.click()
```

```python
# clicking on open source
os = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
os.click()
```

```python
# clicking on trending
tr = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a')
tr.click()
```

```python
# making empty lists to add data in it
url = []
title = []
desc = []
c_count = []
language = []
```

```python
# finding repository title
tle = driver.find_elements(By.XPATH,'//div[@class="Box"]/div[2]/article/h1/a')
for i in tle:
    title.append(i.text)
    
# finding repository description
d = driver.find_elements(By.XPATH,'//div[@class="Box"]/div[2]/article/p')
for i in d:
    desc.append(i.text)
# since there is no description of second repository so using hyphen there
desc.insert(1,'-')

# finding language used in programming
lan = driver.find_elements(By.XPATH,'//div[@class="Box"]/div[2]/article/div[2]/span[1]/span[2]')
for i in lan[0:7]: # since there is no information about language in one repository after 7 repositories
    language.append(i.text)
# adding hyphen where there is no any information about language
language.append('-') 
# adding next repositories languages
for i in lan[7:22]:
    language.append(i.text)
# adding hyphen at right place since there is one more repository which language information is not available
language.append('-')
# adding rest repository languages
for i in lan[22:]:
    language.append(i.text)
```

```python
# finding urls
ur = driver.find_elements(By.XPATH,'//div[@class="Box"]/div[2]/article/h1/a')
for i in ur:
    url.append(i.get_attribute('href'))
```

```python
# importing one more required library
import time
```

```python
# finding contributors count
# using for loop to open all urls one by one
for i in url:
    driver.get(i)
    time.sleep(2)
    
    try:
        no = driver.find_element(By.XPATH,'//div[@class="Layout-sidebar"]/div/div[4]/div/h2/a/span')
        c_count.append(no.text)
        
    except NoSuchElementException:
        c_count.append('-')
```

```python
# checking length of data collected
len(title),len(desc),len(c_count),len(language)
```

```python
# making dataframe
df = pd.DataFrame({'Repository Title':title,'Repository Description':desc,'Contributors count':c_count,'Language used':language})
df
```

```python
# making empty lists to add different data in it
name = []
d_type = []
task = []
a_type = []
instance = []
attribute = []
year = []
```

```python
# finding dataset name
n = driver.find_elements(By.XPATH,'//p[@class="normal"]/b/a')
for i in n:
    name.append(i.text)
    
# finding data type 
ty = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p')
for i in ty[1:]:  # adding only desired data
    d_type.append(i.text)
    
# finding task
tk = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p')
for i in tk[1:]:
    task.append(i.text)
    
# finding attribute type
at = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p')
for i in at[1:]:
    a_type.append(i.text)
    
# finding number of instances
ins = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p')
for i in ins[1:]:
    instance.append(i.text)
    
# finding number of attributes
att = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p')
for i in att[1:]:
    attribute.append(i.text)
    
# finding year
y = driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p')
for i in y[1:]:
    year.append(i.text)
```
